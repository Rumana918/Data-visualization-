```python
#import the dataset
import pandas as pd
df = pd.read_csv('/content/WorldCupMatches.csv')
df
```

```python
#find the number of rows and columns
df.shape
```

```python
#print the each column name
df.columns
```

```python
#print specific range using slicing
df[10:20]
```

```python
#summarization statistics
df.describe()
```

```python
#checking missing values
df.isnull().sum()
```

```python
# Replace missing numerical values with mean/median
df['Attendance'].fillna(df['Attendance'].mean(), inplace=True)
```

```python
#pairplot
import seaborn as sns
sns.pairplot(df)
```

```python
#boxplot
import matplotlib.pyplot as plt
plt.boxplot(df['Attendance'])
plt.show()
```

```python
#heatmap
import seaborn as sns
import matplotlib.pyplot as plt

# Convert date/time columns to datetime objects
# Replace 'Year' with the actual name of your date column
df['Year'] = pd.to_datetime(df['Year'], errors='coerce')

# Extract numerical features for correlation
numerical_df = df.select_dtypes(include=['number'])

corr = numerical_df.corr()
sns.heatmap(corr, annot=True)
plt.show()
```

```python
#barplot
import matplotlib.pyplot as plt
plt.bar(df['Year'], df['Attendance'])
plt.show()
```

```python
#jointplot
import seaborn as sns
sns.jointplot(x='Year', y='Attendance', data=df)
```

```python
#histogram
import matplotlib.pyplot as plt
plt.hist(df['Attendance'])
plt.show()
```

```python
#distribution plot
import seaborn as sns
sns.distplot(df['Attendance'])
```

```python
#scatterplot
import matplotlib.pyplot as plt
plt.scatter(df['Year'], df['Attendance'])
plt.show()
```

```python
#use maximum
df['Attendance'].max()
```

```python
#use minimum
df['Attendance'].min()
```

```python
#use legend()
import matplotlib.pyplot as plt
plt.plot(df['Year'], df['Attendance'], label='Attendance')
plt.legend()
plt.show
```
